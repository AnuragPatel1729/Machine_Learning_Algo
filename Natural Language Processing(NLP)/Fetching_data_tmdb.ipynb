```python
import pandas as pd
import requests
```

```python
pages = 457
```

```python
df = pd.DataFrame()

for page in range(1, pages):
    response = requests.get("https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={}".format(page))
    temp_df = pd.DataFrame(response.json()['results'])
        
    df = pd.concat([df,temp_df],ignore_index=True)
df.head()
```

```python
df.shape
```

```python
movie_data = df[['genre_ids', 'original_title', 'overview']]
```

```python
genre_response = requests.get("https://api.themoviedb.org/3/genre/movie/list?api_key=c45edf1412fd4ed9bb177415aa7f943a&language=en-US")
```

```python
data = genre_response.json()['genres']
```

```python
data
```

```python
genre_data = pd.DataFrame.from_dict(data)
```

```python
genre_data
```

```python
genre_data[genre_data['id']==35]['name']
```

```python
def convert_genre(genre_ids, genre_data):
    genre_names = [genre_data.loc[genre_data['id'] == genre_id, 'name'].iloc[0] for genre_id in genre_ids]
    return ', '.join(genre_names)
```

```python
movie_data['genres'] = movie_data['genre_ids'].apply(lambda x: convert_genre(x, genre_data))
```

```python
movie_data.head()
```

```python
movie_data.drop(columns=['genre_ids'], inplace=True)
```

```python
movie_data.to_csv('TMDB.csv', index=False)
```
